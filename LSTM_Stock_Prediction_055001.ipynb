```python
import numpy as np
import datetime
import pandas as pd

def ag01_str_to_datetime(ag01_s):
    ag01_split = ag01_s.split('-')
    ag01_year, ag01_month, ag01_day = int(ag01_split[0]), int(ag01_split[1]), int(ag01_split[2])
    return datetime.datetime(year=ag01_year, month=ag01_month, day=ag01_day)

def ag01_df_to_windowed_df(ag01_dataframe, ag01_first_date_str, ag01_last_date_str, ag01_n=3):
    ag01_first_date = ag01_str_to_datetime(ag01_first_date_str)
    ag01_last_date = ag01_str_to_datetime(ag01_last_date_str)

    ag01_dates = []
    ag01_X, ag01_Y = [], []

    ag01_current_date = ag01_first_date

    while ag01_current_date <= ag01_last_date:
        ag01_window = []
        ag01_temp_date = ag01_current_date

        # Find the values for the previous n valid trading days
        for i in range(ag01_n):
            ag01_temp_date -= datetime.timedelta(days=1)
            while ag01_temp_date >= ag01_dataframe.index[0]:
                ag01_temp_date_dt = pd.to_datetime(ag01_temp_date)
                if ag01_temp_date_dt in pd.to_datetime(ag01_dataframe.index):
                    ag01_window.insert(0, ag01_dataframe.loc[ag01_temp_date_dt])
                    break
                ag01_temp_date -= datetime.timedelta(days=1)
            else: #break out if we are out of dates.
                break

        # Check if we have enough data for the window
        if len(ag01_window) == ag01_n and pd.to_datetime(ag01_current_date) in pd.to_datetime(ag01_dataframe.index):
            ag01_X.append(ag01_window)
            ag01_Y.append(ag01_dataframe.loc[pd.to_datetime(ag01_current_date)])
            ag01_dates.append(ag01_current_date)

        ag01_current_date += datetime.timedelta(days=1)

    ag01_ret_df = pd.DataFrame({})
    ag01_ret_df['Target Date'] = ag01_dates

    ag01_X = np.array(ag01_X)
    for ag01_i in range(0, ag01_n):
        ag01_ret_df[f'Target-{ag01_n-ag01_i}'] = ag01_X[:, ag01_i]

    ag01_ret_df['Target'] = ag01_Y

    return ag01_ret_df
```

```python
import numpy as np

def ag01_windowed_df_to_date_X_y(ag01_windowed_dataframe):
    ag01_df_as_np = ag01_windowed_dataframe.to_numpy()

    ag01_dates = ag01_df_as_np[:, 0]

    ag01_middle_matrix = ag01_df_as_np[:, 1:-1]
    ag01_X = ag01_middle_matrix.reshape((len(ag01_dates), ag01_middle_matrix.shape[1], 1))

    ag01_Y = ag01_df_as_np[:, -1]

    return ag01_dates, ag01_X.astype(np.float32), ag01_Y.astype(np.float32)
```

```python
# Model for Nifty
from tensorflow.keras.models import Sequential
from tensorflow.keras.optimizers import Adam
from tensorflow.keras import layers
from tensorflow.keras.callbacks import ReduceLROnPlateau

ag01_model_nifty = Sequential([
    layers.Input((3, 1)),
    layers.LSTM(128, return_sequences=True),  # Increased units, return sequences for stacked LSTM
    layers.LSTM(64),  # Added another LSTM layer
    layers.Dense(32, activation='relu'),
    layers.Dropout(0.2),  # Added dropout for regularization
    layers.Dense(32, activation='relu'),
    layers.Dense(1)
])

# Learning rate scheduler
lr_scheduler = ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=5, min_lr=0.0001)

ag01_model_nifty.compile(loss='mse',
                    optimizer=Adam(learning_rate=0.001),
                    metrics=['mean_absolute_error'])

ag01_model_nifty.fit(ag01_X_train_nifty_scaled, ag01_y_train_nifty_scaled,
                    validation_data=(ag01_X_val_nifty_scaled, ag01_y_val_nifty_scaled),
                    epochs=100,
                    callbacks=[lr_scheduler])
```

```python
# Model for HUL
ag01_model_hul = Sequential([layers.Input((3, 1)),
                          layers.LSTM(64),
                          layers.Dense(32, activation='relu'),
                          layers.Dense(32, activation='relu'),
                          layers.Dense(1)])

ag01_model_hul.compile(loss='mse',
                    optimizer=Adam(learning_rate=0.001),
                    metrics=['mean_absolute_error'])

ag01_model_hul.fit(ag01_X_train_hul, ag01_y_train_hul, validation_data=(ag01_X_val_hul, ag01_y_val_hul), epochs=100)
```

```python
# For Nifty
ag01_train_predictions_nifty = ag01_model_nifty.predict(ag01_X_train_nifty).flatten()

# Reshape to 2D for inverse_transform
ag01_train_predictions_nifty_2d = ag01_train_predictions_nifty.reshape(-1, 1)

# Inverse transform
ag01_train_predictions_nifty_original = ag01_scaler_nifty.inverse_transform(ag01_train_predictions_nifty_2d)

# Reshape back to 1D for plotting (if necessary)
ag01_train_predictions_nifty_original = ag01_train_predictions_nifty_original.flatten()

plt.plot(ag01_dates_train_nifty, ag01_train_predictions_nifty_original)
plt.plot(ag01_dates_train_nifty, ag01_y_train_nifty)
plt.legend(['Training Predictions Nifty', 'Training Observations Nifty'])

```
